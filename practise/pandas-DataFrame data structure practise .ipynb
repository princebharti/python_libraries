## GroupBy operation on dataframe

```python
import pandas as pd
data=pd.read_csv("mlbsalaries.csv")
data.head(50)
```

```python
grouped=data.groupby(by=['Year'])
```

```python
grouped
```

```python
for k,group in grouped:
    print("the key is :",k)
    print(group)
    print("\n\n\n")
```

```python
gen=(group.sort_index(ascending=False,by=['Salary'])[:1] for k,group in grouped)
```

```python
gen
```

```python
newdf=pd.DataFrame()

for line in gen:
    newdf=newdf.append(line)
    
```
